```python
import os
import google.generativeai as genai
from dotenv import load_dotenv

# 1. Cấu hình
load_dotenv()
genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

if __name__ == "__main__":
    # 2. Khởi tạo model
    model = genai.GenerativeModel('gemini-2.5-flash')

    # 3. Prompt yêu cầu tính toán trung bình nhân (Geometric Mean)
    prompt = """
    Giải thích ngắn gọn công thức Trung bình nhân (Geometric Mean). 
    Sau đó tính trung bình nhân của hai số 5 và 20. 
    Trình bày các bước tính toán rõ ràng.
    """

    # 4. Gửi request
    response = model.generate_content(prompt)

    # 5. Xuất kết quả
    print("--- KẾT QUẢ LAB 3.1 (GEMINI) ---")
    print(response.text)
```
