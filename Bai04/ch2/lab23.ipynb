# Lab 2.3 - Code Generation with Gemini API
Using Google Generative AI (Gemini) to generate Python code

````python
# Cell 1: Import Libraries
import os
import re
from dotenv import load_dotenv
import google.generativeai as genai
````

````python
# Cell 2: Initialize Model and Configuration
# Load API Key from .env file
load_dotenv(os.path.join(os.path.dirname(os.path.abspath('.')), '.env'))
api_key = os.getenv("GEMINI_API_KEY")

# Configure Gemini API
genai.configure(api_key=api_key)

# Initialize model with high temperature for creativity
generation_config = genai.types.GenerationConfig(
    temperature=1.0,
    candidate_count=2
)

model = genai.GenerativeModel(
    'gemini-2.5-flash',
    generation_config=generation_config
)

# Define prompts
USER_PROMPT = """
def print_fibonacci_sequence(n: int) -> None:
"""

SYSTEM_PROMPT = "You will be provided with a Python function signature. Your task is to implement the function. Return code only."
````

````python
# Cell 3: Execute and Print Results
# Helper function to add instruction comment
def get_code_with_instructions(code: str) -> str:
    """Add a comment to the code for specific code completion instruction"""
    return code + "\n# Complete this code"

# Prepare the full prompt
full_prompt = f"{SYSTEM_PROMPT}\n\n{get_code_with_instructions(USER_PROMPT)}"

# Generate content
response = model.generate_content(full_prompt)

# Extract and display code suggestions
for i, candidate in enumerate(response.candidates):
    output = candidate.content.parts[0].text
    # Extract code from markdown blocks if present
    code_suggestion = re.sub(r"(.*?)```python(.*?)```(.*)", r"\2", output, flags=re.DOTALL).strip()
    if not code_suggestion:
        code_suggestion = output
    print(f"Output {i + 1}:")
    print(code_suggestion)
    print("\n" + "="*50 + "\n")
````
