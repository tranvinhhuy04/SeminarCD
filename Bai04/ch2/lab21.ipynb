# Lab 2.1 - FizzBuzz Problem with Gemini API
Using Google Generative AI (Gemini) to explain the FizzBuzz problem

```python
# Cell 1: Import Libraries
import os
from dotenv import load_dotenv
import google.generativeai as genai
```

```python
# Cell 2: Initialize Model and Configuration
# Load API Key from .env file
load_dotenv(os.path.join(os.path.dirname(os.path.abspath('.')), '.env'))
api_key = os.getenv("GEMINI_API_KEY")

# Configure Gemini API
genai.configure(api_key=api_key)

# Initialize the model
model = genai.GenerativeModel('gemini-2.5-flash')
```

```python
# Cell 3: Execute and Print Results
# Send request to Gemini API
response = model.generate_content("What is the FizzBuzz problem?")

# Print the response
print(response.text)
```
