# Lab 2.2 - Two Sum Problem with Gemini API
Using Google Generative AI (Gemini) to solve the Two Sum problem

```python
# Cell 1: Import Libraries
import os
from dotenv import load_dotenv
import google.generativeai as genai
```

```python
# Cell 2: Initialize Model and Configuration
# Load API Key from .env file
load_dotenv(os.path.join(os.path.dirname(os.path.abspath('.')), '.env'))
api_key = os.getenv("GEMINI_API_KEY")

# Configure Gemini API
genai.configure(api_key=api_key)

# Initialize the model with specific generation configuration
generation_config = genai.types.GenerationConfig(
    temperature=0.2,
    max_output_tokens=100,
    candidate_count=1
)

model = genai.GenerativeModel(
    'gemini-2.5-flash',
    generation_config=generation_config
)
```

```python
# Cell 3: Execute and Print Results
# Create messages for system and user content
response = model.generate_content([
    "You are a hiring manager at a tech company.",
    "What is the Two Sum problem?"
])

print(response.text)
```
