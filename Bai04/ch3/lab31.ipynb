# Lab 3.1 - Geometric Mean of Two Numbers
Calculate the geometric mean of two numbers

```python
# Cell 1: Import Libraries
# No external imports needed for this operation
```

```python
# Cell 2: Initialize Model and Configuration
def get_geometric_mean_of_two_numbers(
        a: float,
        b: float,
) -> float:
    """Calculate the geometric mean of two numbers"""
    return pow(a * b, 1 / 2)
```

```python
# Cell 3: Execute and Print Results
num1: float = 5.0
num2: float = 20.0
result = get_geometric_mean_of_two_numbers(num1, num2)
print(f"Geometric mean of {num1} and {num2}: {result}")
```
