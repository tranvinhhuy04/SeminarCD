Calculate the geometric mean of multiple numbers with validation

```python
# Cell 2: Initialize Model and Configuration
def get_geometric_mean(*nums):
    """
    Calculate the geometric mean of multiple numbers.
    
    Args:
        *nums: Variable length argument list of numbers
        
    Returns:
        float: The geometric mean of the input numbers
        
    Raises:
        ValueError: If no numbers are provided
        TypeError: If any number is not int or float
    """
    if not nums:
        raise ValueError("At least one number is required")
    for num in nums:
        if not isinstance(num, (int, float)):
            raise TypeError("All numbers must be integers or floats")
    product = 1
    for num in nums:
        product *= num
    return pow(product, 1 / len(nums))
```

```python
# Cell 3: Execute and Print Results
# Test with different sets of numbers
test_cases = [
    (2, 8),
    (1, 2, 4, 8),
    (5.0, 10.0, 15.0),
]

for nums in test_cases:
    result = get_geometric_mean(*nums)
    print(f"Geometric mean of {nums}: {result}")

# Test error handling
try:
    get_geometric_mean()
except ValueError as e:
    print(f"\nError caught: {e}")

try:
    get_geometric_mean(1, "string", 3)
except TypeError as e:
    print(f"Error caught: {e}")
```
