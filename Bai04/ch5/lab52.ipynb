# Lab 5.2 - Singleton Pattern with Unit Tests
Testing the Singleton pattern implementation

```python
# Cell 1: Import Libraries
from dataclasses import dataclass
from unittest import TestCase, main
```

```python
# Cell 2: Initialize Model and Configuration
class Singleton(type):
    """Metaclass implementing the Singleton pattern"""
    _instances = {}

    def __call__(cls, *args, **kwargs):
        if cls not in cls._instances:
            cls._instances[cls] = super(Singleton, cls).__call__(*args, **kwargs)
        return cls._instances[cls]

@dataclass
class Environment(metaclass=Singleton):
    """Environment configuration using Singleton pattern"""
    name: str = 'Production'  # This is the default environment
    version: int = 1

class TestSingleton(TestCase):
    """Unit tests for Singleton pattern"""
    
    def test_singleton_should_return_same_instance(self):
        """Test that Singleton returns the same instance"""
        @dataclass
        class A(metaclass=Singleton):
            a: int = 0

        self.assertEqual(A(2), A())
```

```python
# Cell 3: Execute and Print Results
# Run the test
import unittest

# Create a test suite
loader = unittest.TestLoader()
suite = loader.loadTestsFromTestCase(TestSingleton)

# Run the tests
runner = unittest.TextTestRunner(verbosity=2)
result = runner.run(suite)

# Verify Singleton behavior
print("\n" + "="*80)
print("Testing Singleton Pattern:")
print("="*80)

env1 = Environment()
env2 = Environment(name='Development', version=2)

print(f"env1: {env1}")
print(f"env2: {env2}")
print(f"env1 is env2: {env1 is env2}")
print(f"env1 == env2: {env1 == env2}")
```
