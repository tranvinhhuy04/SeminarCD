# Lab 5.1 - Docstring Generation with Gemini API
Using Google Generative AI (Gemini) to generate Google Style docstrings

```python
# Cell 1: Import Libraries
import os
import inspect
from dotenv import load_dotenv
import google.generativeai as genai
```

```python
# Cell 2: Initialize Model and Configuration
# Load API Key from .env file
load_dotenv(os.path.join(os.path.dirname(os.path.abspath('.')), '.env'))
api_key = os.getenv("GEMINI_API_KEY")

# Configure Gemini API
genai.configure(api_key=api_key)

# Initialize model
model = genai.GenerativeModel('gemini-2.5-flash')

# Define prompts
SURROUND = "You will be provided with a Python function enclosed with {{{ Function }}}."
SINGLE_TASK = "Your task is to generate Google Style docstring for it."

class Singleton(type):
    """Metaclass for implementing the Singleton pattern"""
    _instances = {}

    def __call__(cls, *args, **kwargs):
        if cls not in cls._instances:
            cls._instances[cls] = super(Singleton, cls).__call__(*args, **kwargs)
        return cls._instances[cls]

def get_user_prompt(func: callable) -> str:
    """Generate user prompt for docstring generation"""
    return f"""
    FUNCTION: {{{{ {inspect.getsource(func)} }}}}

    GOOGLE STYLE DOCSTRING:
    """
```

```python
# Cell 3: Execute and Print Results
system_prompt = f"{SURROUND} {SINGLE_TASK}"
user_prompt = get_user_prompt(Singleton.__call__)

response = model.generate_content(f"{system_prompt}\n{user_prompt}")

print("Generated Docstring:")
print("=" * 80)
print(response.text)
```
